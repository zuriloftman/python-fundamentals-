```python
titanic["binned_ages"] = pd.cut(titanic["age"], bins = 10)

titanic.head()

#binning categorises data. 
#putting it in a range is easier for us to work out what we want to do with it
#and how we value it
#for age data the range is 0-8 and 24-32
#Note the bracket denotes exclusion of that number 
#and the square bracket denotes inclusion of that number.

```

```python
titanic["binned_ages"].unique()
```

```python
#we can also pass our own values to determine what the edges of 
#categories are instead of python computing the approximate ones

titanic["binned_ages2"]  = pd.cut(titanic["age"],  # Data to cut
                                  bins=np.arange(start=0,
                                                 stop=(titanic["age"].max() + 1) , # Remember stop is exclusive!
                                                 step=10))

titanic.head()

#the code means we don't know the maximum value for the column before 
#writing the code
#with: stop=(titanic[“age”].max() + 1)
#I use brackets to enforce the order of operations. 
#If I didn’t add 1 to the stop, the entry at the maximum 
#value would read NaN, which I want to avoid.
```

```python
#We can also add labels to our categories. 
#This time rather than displaying the bin edges it will display 
#the text strings we specify.

#This is passed as a list to the parameter labels= to the 
#pd.cut() function.

titanic["binned_ages3"] = pd.cut(titanic["age"],  # Data to cut
                                 bins=np.arange(start=0,
                                                        stop=(titanic["age"].max() + 1),
                                                step=10),
                                 labels=["0 – 10", "11-20", " 21 – 30",
                                         "31 – 40", "41 – 50", "51 – 60",
                                         "61 – 70", "71 - 80"])

titanic.head()

#these bands are approximate, e.g someone with an age of 20.2 
#will go into the band labelled “21-30”. 
#Integers were chosen as it’s an easier read, 
#and most ages after 1 are whole numbers.

#Personal Summary:
#binning categorises data 
#putting data in a range is easier for us to work out what we want 
#to do with it and hiw we value it 
```

```python
#If we want to return more than one aggregation at once to the columns
#we’ve selected, there is the .agg() method. 

#This takes a dictionary
#where the key is our column and the value is the aggregation method
#we wish to apply as a string.


animalclass_sum = animals.groupby(by="AnimalClass").agg({"IncidentNominalCost(£)": "sum",
                                                         "PumpHoursTotal": "mean"})

animalclass_sum
```

```python
#If we want to apply more than one aggregation to a column 
#we can pass a list to the values of the dictionary. 
#This requires us to use the numpy methods which are very similar 
#to the methods we have used previously, so np.sum , np.mean etc.

#Group the animals dataset by the variable Borough and obtain the 
#following summary statistics:

#•The variance and standard deviation of the IncidentNominalCost(£) 
#column.
#•The mean of the PumpCount column.
#•The sum of the PumpHoursTotal column.

animals.aggregates = animals.groupby(by = "Borough").agg ({"IncidentNominalCost(£)": [np.var, np.std],
                                                          "PumpCount": "mean",
                                                          "PumpHoursTotal": "sum"})

animals.aggregates
```
