```python
#Filling in null values 
#making a copy df

pun_animals_fill = pun_animals.copy()
```

```python
#telling to fill missing values (zero) with the mean 

missing_puns = round(number=pun_animals_fill["steps_per_day"].mean(), ndigits = 0)
```

```python
#filling missing (null) values as the mean value of the steps_per_day 

pun_animals_fill["steps_per_day"].fillna(value=missing_puns, inplace = True)
```
